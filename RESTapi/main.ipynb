# Aula 03: Introduzindo API's

Uma API (Application Programming Interface) é uma aplicação desenvolvida em uma linguagem de programação para servir dados e informações para outras aplicações, independente se foram construídas na mesma linguagem ou com o mesmo paradigma

Ela opera como um serviço, recebendo e retornando sados para aqueles que o requisitam.

Existem diversos tipos de API, porém os que mais se destacam são: **SOAP Request**, **GraphQL** e **REST**.

# Inicializando o projeto

````javascript

    npm init    // cria o arquivo package.json que gerencia os pacotes e dependências
    npm init -y // pula as etapas de input do init

    // para facilitar o desenvolvimento da API, iremos utilizar a seguinte configuração

    "type": "module"; // indica o código está utilizando módulos ES

    // Antes de iniciar a API, deve-se configurar o projeto para aceitar o uso de TS
    npm i typescript @types/node tsx -D

    // agora roda o comando abaixo para criar o arquivo de configuração do ts, para assim gerenciá-lo
    npx tsc --init

    // Depois de selecionar a versão, precisa configurar o necessário para a API rodar corretamente juntamente ao TS
    "scripts": {
        "test": "echo \"Error: no test specified\" && exit 1",
        "dev": "npx tsx watch src/server.ts"
    },
````

### Express
Como framework, o Express será utilizado para fornecer a possibilidade de criar um servidor HTTP com o JS. Para instalar o exprees faça:
````javascript
    npm i express @types/express
````

Para configurar o servidor HTTP, iremos utilizar uma base de código que inicialirá o servidor em "./src/server.ts"
````javascript
    import express from 'express':

    const app = express();
    const port = 8080;

    app.listen(port, () => console.log(`Acesse: htttp://localhost:${port}/`))
````
A partir deste ponto, todas as resquisições podem ser utilizada dentro de server.ts

### Exemplo de GET
````javascript

    app.get('/getTest', (req, res) => {
        res.send('GET: Requisição recebida com sucesso!');
    })
````
