# **Aula 01** 

O MongoDB é um banco de dados NoSQL de código aberto, orientado a documentos e de alta performance, que armazena informações em formato JSON (BSON) em vez de tabelas relacionais. Ele é ideal para aplicações modernas que exigem flexibilidade no esquema, escalabilidade horizontal e rápido desenvolvimento. 

Principais Características e Benefícios:
- **Modelo Orientado a Documentos**: Dados relacionados são armazenados juntos em documentos (semelhantes a objetos JSON), facilitando a modelagem.

- **Esquema Flexível**: Não requer uma estrutura rígida, permitindo que os campos mudem conforme a aplicação evolui.

- **NoSQL/Não Relacional**: Diferente de bancos SQL (como MySQL), não utiliza tabelas e chaves estrangeiras complexas.

- E**scalabilidade e Desempenho**: Projetado para fácil particionamento (sharding) e replicação, suportando alto volume de dados e tráfego.

- **BSON (Binary JSON)**: Formato binário que armazena dados de forma rápida e eficiente, suportando diversos tipos de dados.
